1. An√°lisis Exploratorio:
Describir la serie de tiempo y visualizarla.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carga el archivo CSV en un DataFrame de Pandas
df = pd.read_csv('daily-total-female-births.csv')

# Aseg√∫rate de que la columna 'Date' se interprete como fechas
df['Date'] = pd.to_datetime(df['Date'])

# Muestra informaci√≥n b√°sica sobre el DataFrame
print(df.info())

# Muestra estad√≠sticas descriptivas de la columna 'Births'
print(df['Births'].describe())

# Configura el tama√±o de la figura
plt.figure(figsize=(12, 6))

# Grafica la serie de tiempo
sns.set_style("darkgrid")
plt.plot(df['Date'], df['Births'], label='Nacimientos diarios', color='b')

# Configura etiquetas y t√≠tulo
plt.xlabel('Fecha')
plt.ylabel('Nacimientos')
plt.title('Nacimientos Diarios de Mujeres')

# Rota las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45)

# Muestra la leyenda
plt.legend()

# Muestra el gr√°fico
plt.show()
```
